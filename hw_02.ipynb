```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

headerList = ['범죄율','주거지 비율','비소매 업무 비율','찰스 강 가장 여부','질소 산화물 농도','평균 방 갯수','1940년 이전 건물 비율','고용 센터까지 가중 거리',
'고속도로 접근성 지수','부동산 세율','학생 대 교사 비율','블랙 인구 비율','인구 하위 계층 비율', '주택의 중앙값']

data = pd.read_csv('boston.csv', index_col=False)

data = pd.DataFrame(data)
data.columns = headerList
data
```

```python
data.shape, data.info, data.dtypes
```

```python
jutaek = data['주택의 중앙값']
```

```python
print(f'''
1. 주택 가격의 평균 : {jutaek.mean()}
2. 주택 가각의 중앙값 : {jutaek.median()}
3. 주택 가격의 표준 편차 :{jutaek.std()}
4. 주택 가격의 최대값 :{jutaek.max()}
5. 주택 가격의 최소값 :{jutaek.min()}
6. 주택 가격의 최빈값 :{jutaek.mode()[0]}
7. 주택 가격의 분포 : {jutaek.std()**2}
''')
```

```python
plt.figure(figsize = (10, 6))
plt.hist(jutaek, bins = 11, range = (0, 55), color = 'tomato', width=4)
plt.title('단위($1000 dollar)', loc = 'left')
plt.title('<보스턴 주택 가격 분포>', size = 15)
plt.xticks(np.linspace(0, 55, 12))
plt.xlabel('주택 가격', size = 15)
plt.ylabel('주택 개수', size = 15)
plt.show()
```
