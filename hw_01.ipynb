```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

data = pd.read_csv('CARD_SUBWAY_MONTH_202312.csv', index_col=False)

data = pd.DataFrame(data)
data
```

```python
data['총승객수'] = data['승차총승객수'] + data['하차총승객수']
data['노선역명'] = data['노선명'] + ' ' + data['역명']
data
```

```python
data[['노선명','역명']].iloc[0]['노선명']
```

```python
list(data['역명'].unique()), len(list(data['역명'].unique()))
stationList = list(data['역명'].unique())
stationList
```

```python
stationPeopleList = []


```

```python
for s in stationList:
    stationDF = data['역명'] == s
    staDF = data[stationDF]
    # print(staDF)
    stationPeopleList.append(staDF['총승객수'].sum(axis=0))
    print(staDF['총승객수'].sum(axis=0))
    print(stationPeopleList)

staDF

staDict = dict(zip(stationList,stationPeopleList))
staDict


# for k,v in staDict.items():
#     print(k,v)
#     aList = [v, v+10]
#     staDict[k] = aList
# staDict
```

```python
staDict
```

```python
staDF = pd.Series(staDict)
staDF
```

```python
staNP = np.array(staDF)
staNP
```

```python
count = 1
print(f'각 지하철 역별 평균 이용객 수')
for k, v in staDict.items():
    print(f'{count:>3} : {k:>15}역 - {v//31:,}명')
    count += 1
```

```python
deviation = staNP - np.mean(staNP)
deviation

count = 1
print(f'2. 각 지하철 역별 이용객 수의 표준 편차')
for k, v in staDict.items():
    print(f'{count:3} : {k:>10}역 : {deviation[count-1]:,}')
    count += 1
```

```python
count = 1
maxV = 0
maxK = ''

print(f'3. 가장 많은 이용객이 있는 역의 이용객 수')
for k, v in staDict.items():
    if v > maxV:
        maxV = v
        maxK = k
    count += 1
print(f'{maxV:,}명')
```

```python
count = 1
minV = 99999999999999999999
minK = ''

print(f'4. 가장 적은 이용객이 있는 역의 이용객 수')
for k, v in staDict.items():
    if v < minV:
        minV = v
        minK = k
    count += 1
print(f'{minV:,}명')
```

```python
count = 1
maxV = 0
maxK = ''

print(f'5. 이용객 수가 가장 많은 역의 이름')
for k, v in staDict.items():
    if v > maxV:
        maxV = v
        maxK = k
    count += 1
print(f'{maxK}역')
```

```python
count = 1
minV = 99999999999999999999
minK = []

print(f'6. 이용객 수가 가장 적은 역의 이름')
for k, v in staDict.items():
    if v < minV:
        minV = v
        minK.append(k+'역')
    count += 1

print(",".join(minK))
```
