```python
df1, df2, df3, df4 = df['sepal_length'], df['sepal_width'], df['petal_length'], df['petal_width']
```

```python
df['species'].unique()
```

```python
speciesList = ['setosa', 'versicolor', 'virginica']
```

```python
def spec(species):
    df5 = df['species']
    df5_1 = df5 == species
    dfIris = df[df5_1]
    # print(dfIris)
    return dfIris
print(dfIris.columns[0])
```

```python
setosa = spec(speciesList[0])
versicolor = spec(speciesList[1])
virginica = spec(speciesList[2])
```

```python
spec(speciesList[0])['sepal_length']
```

```python
for i in range(len(speciesList)):
    spec(speciesList[i])
```

```python
def histo(DF, title, xlabel, ylabel):
    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot(111)
    a,b,c = ax.hist(DF, bins=15)
    ax.set_title(title, size=20)
    ax.set_xticks(b)
    ax.set_yticks(a)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

histo(df1, 'Sepal_length Histogram', 'sepal_length', 'number')
histo(df2, 'Sepal_widgth Histogram', 'sepal_widgth', 'number')
histo(df3, 'Petal_length Histogram', 'petal_length', 'number')
histo(df4, 'Petal_widgth Histogram', 'petal_widgth', 'number')
```

```python
fig = plt.figure(figsize=(7,7))

def boxpl(DF, title, xlabel, ylabel,i):
    ax = fig.add_subplot(i)
    ax.boxplot(DF)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.hist

fig = plt.figure(figsize=(7,7))
plt.suptitle(speciesList[0].upper())
plt.hist
boxpl(setosa['sepal_length'], 'Sepal_length Histogram', 'sepal_length', 'number',221)
boxpl(setosa['sepal_width'], 'Sepal_widgth Histogram', 'sepal_width', 'number',222)
boxpl(setosa['petal_length'], 'Petal_length Histogram', 'petal_length', 'number',223)
boxpl(setosa['petal_width'], 'Petal_widgth Histogram', 'petal_width', 'number',224)
```
